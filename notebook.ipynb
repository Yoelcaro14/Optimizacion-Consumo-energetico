## REGISTRO CONSUMO

```python
# Carga datos
df= pd.read_excel('C:/Users/yoe_1/OneDrive/Escritorio/CARPETA_DOCUMENTOS/Medidor_luz/REGISTRO DE MEDIDOR.xlsx',sheet_name='REGISTRO',usecols='A:I',header=3)
df.columns = df.columns.str.strip()
df.head()
```

```python
# Limpieza de datos
df['MES']=df['MES2']
df = df.drop('MES2',axis=1)                     # MES           - Se establece solo una columna MES
df = df.drop(df[df['VAR. LECTURA']==0].index)   # VAR. LECTURA	- Se eliminan valores nulos irrelevantes
df.head()
```

```python
# Se completa Val. faltantes en lectura
ind_lectura_na = df[df['ITEM']=='ANDY'].index.tolist()

for i in ind_lectura_na:
    df.loc[i,'LECTURA']=df.loc[i-3,'LECTURA']-df.loc[i-2,'LECTURA']-df.loc[i-1,'LECTURA']
```

```python
# Tabla oficial del registro de consumo por individuo
df_indi = df.drop(df[df['ITEM']=='TOTAL'].index).reset_index(drop=True)
df_indi = df_indi.drop('KWh',axis=1)
df_indi = df_indi.sort_index(ascending=False).reset_index(drop=True)    # Ordenando indices
df_indi.head(20)
```
